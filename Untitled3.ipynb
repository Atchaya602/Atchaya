```python
# Step 1: Import libraries
import pandas as pd
import json

# Step 2: Load the uploaded JSON file
with open('users.json', 'r') as file:
    data = json.load(file)

# Step 3: Convert to pandas DataFrame
df = pd.DataFrame(data)

# Step 4: Preview the data
df.head()
```

```python
df[df['address'].apply(lambda x: x['city']) == 'South Christy'][['name', 'email']]
```

```python
df['company_name'] = df['company'].apply(lambda x: x['name'])
df['company_name'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Count users per city
df['city'] = df['address'].apply(lambda x: x['city'])
df['city'].value_counts().plot(kind='bar', color='skyblue')

plt.title('Users per City')
plt.xlabel('City')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
